```python
from pandas import Series, DataFrame

data = {'state': ['Ohio','Ohio', 'Ohio', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001,2002],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9]}
frame = DataFrame(data) # 딕셔너리 데이터를 DataFrame 객체 생성
frame2 = DataFrame(data, columns = ['year', 'state', 'pop', 'debt'],
                   index=['one','two','three', 'four', 'five'])
frame2
```

```python
import json
import glob

rootdir = '/content/drive/MyDrive/2025_SeSAC/python_programming/일반남여_일반통합06_F_1524374974_25_수도권_실내/'
flist = glob.glob(rootdir+'/*.json') # 파일 경로에서 특정 패턴에 맞는 파일 목록을 가져올 때

stt_list = []
thema_list = []
for fname in flist[:10]:
  # 파일을 오픈한 내용을 넣어야 함
  with open(fname) as f:
    json_data = json.load(f) # json의 구조로 되어있는 데이터 읽음

    fileNm = json_data['발화정보']['fileNm']
    stt = json_data['발화정보']['stt']
    thema = json_data['대화정보']['convrsThema']

    stt_list.append({'file':fileNm, 'stt':stt})
    thema_list.append({'stt':stt, 'thema':thema})

df_tss = DataFrame(stt_list,columns = ['file', 'stt'])
df_thema = DataFrame(thema_list,columns = ['stt', 'thema'])

df_tss.to_csv('stt.csv')
df_thema.to_csv('thema.csv')
print("저장완료")
```
