```python
for a in [1, 2, 3]:
  for b in [10, 20, 30]:
    if a*10 != b: # b//a != 10
      print(b)
  print(a)
print("done")
```

```python
i = 0
a = 0
while True:
  print(a)
  if i < 5:
    i += 1
  else:
    break
print("done")
```

```python
i = 0
a = 0
while i < 5:
  print(a)
  i += 1
print("done")
```

```python
for a in [1, 2, 3, 4]:
  i = a
  while i > 0:
    print(a)
    i -= 1 #+
```
