```python
# 1. space 기준으로 나누고
# 2. 3글자 이하 제외
# 3. 단어 빈도가 많은 상위 10개 출력
# 강사님 풀이
def top_ten_word(msg):
  top_ten = []
  word_cnt = {}
  for w in msg.split(): # split하면서 for문 돌려서 한번에 실행하기
    nw = w.lower() # 대소문자 따로 카운팅 하는걸 방지하기 위해 소문자로 만들기
    nw = nw.replace(".","")
    nw = nw.replace(",","")
    nw = nw.replace('"',"")
    nw = nw.replace("'s","")
    nw = nw.replace(":","")
    nw = nw.replace(";","")
    nw = nw.replace("?","")
    nw = nw.replace("!","")
    if not nw.isalpha(): # 알파벳이 아닌 다른 문자가 들어간 단어만 출력
      print(nw)
      continue # 해당 구역의 반복문 탈출 -> 여기에서는 바로 상위에 있는 for문으로 감

    if len(nw) <= 3:
      continue
    word_cnt[nw] = word_cnt.get(nw, 0) + 1
  print("word_cnt :",word_cnt)

  cnt_word = {}
  for w in word_cnt:
    cnt_word[word_cnt[w]] = cnt_word.get(word_cnt[w], []) + [w]
  print("cnt_word :",cnt_word)

  for cnt in sorted(cnt_word.keys(), reverse=True): # 키를 가져오는 것과 정렬 하는걸 한번에 실행
    if len(top_ten) <= 10:
      top_ten.extend(cnt_word[cnt]) # cnt_word[cnt]의 요소가 여러개이기 때문에 extend 사용
    else: # else break를 사용하는 이유는 for문을 끝까지 도는걸 막기 위해
      break
  print("top_ten :", top_ten)

  # .items() 튜플로 묶어서 리스트로 반환

  return top_ten


top_ten_word(msg)
```

```python
# Counter 클래스 생성해 보기
class Counter:
  def reset(self):
    self.count = 0
  def increment(self):
    self.count += 1
  def get(self):
    return self.count

# 객체 생성해 보기
a = Counter()

a.reset()
a.increment()
a.get()
```

```python
# 생성자 : 기본 변수나 먼저 실행되야 하는 코드를 정의 하기 def __init__(self): /  선언할 때 반드시 있어야 하는 값
class Counter:
  def __init__(self):
    self.__count = 0 # __는 외부에 공개하지 말아야 하는 변수임을 의미
  def reset(self):
    self.__count = 0
  def increment(self):
    self.__count += 1
  def get(self):
    return self.__count

b = Counter() # 인스턴스가 생성될때 init 실행되면서 필요한 변수 정의 됨

b.increment()
b.get()
```

```python
# 메소드 재정의
# 내 풀이
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def setName(self, name):
    self.__name = name
  def setAge(self, age):
    self.__age = age
  def getName(self):
    return self.name
  def getAge(self):
    return self.age

class Student(Person):
  def __init__(self, name, age, major=False):
    super().__init__(name, age)
    self.major = major

  def setMajor(self, major):
    self.major = major

  def __str__(self):
    if self.major:
      return "[name : %s, age : %d, major : %s]" % (self.name, self.age, self.major)
    else:
      return "[name : %s, age : %d]" % (self.name, self.age)

s1 = Student("song", 22, "a")
s2 = Student("hee", 11)
print("s1 :", s1)
print("s2 :", s2)
```

```python
# 강사님 풀이
class Person:
  def __init__(self, name, age):
    self.__name = name # get메서드가 있으면 외부에서 값을 변경할 수 없도록 __를 붙임
    self.__age = age

  def setName(self, new_name):
    self.__name = new_name
  def setAge(self, new_age):
    self.__age = new_age
  def getName(self):
    return self.__name
  def getAge(self):
    return self.__age

  def __str__(self):
    return f'name:{self.__name}, age:{self.__age}'

a = Person("Kim", 30)
b = Person("Lee", 25)
print(a, b)

class Student(Person):
  def __init__(self, name, age, major):
    super().__init__(name, age)
    self.__major = major

  def setMajor(self, new_major):
    self.__major = new_major
  def getMajor(self):
    return self.__major

  def __str__(self):
    # return f'name:{self.__name}, age:{self.__age}, major:{self.__major}' #  name, age 상속받은 곳에 있음
    # return f'name:{super().__name}, age:{super().__age}, major:{self.__major}' # 변수를 __ 히든으로 만들어서 바로 사용 못함.
    return f'name:{super().getName()}, age:{super().getAge()}, major:{self.__major}'

c = Student("Ko", 30, "computer science")
print(c)
c.setName("ku")
print(c)
```

```python
# 1. Person 이름, 나이 둘다 같은 것만 같은 사람임
# 2. Student 이름, 나이, 전공 셋다 같은 것만 같은 사람임
class Person:
  def __init__(self, name, age):
    self.__name = name
    self.__age = age

  def setName(self, new_name):
    self.__name = new_name
  def setAge(self, new_age):
    self.__age = new_age
  def getName(self):
    return self.__name
  def getAge(self):
    return self.__age

  def __str__(self):
    return f'name:{self.__name}, age:{self.__age}'

  # 1
  def __eq__(self, personB): return self.__name == personB.__name and self.__age == personB.__age


a = Person("Kim", 30)
b = Person("Lee", 25)
c = Person("Kim", 30)
print(f'a = {a}\nb = {b}\nc = {c}')
print("a == b",a == b)
print("a == c",a == c)

class Student(Person):
  def __init__(self, name, age, major):
    super().__init__(name, age)
    self.__major = major

  def setMajor(self, new_major):
    self.__major = new_major
  def getMajor(self):
    return self.__major

  def __str__(self):
    return f'name:{super().getName()}, age:{super().getAge()}, major:{self.__major}'

  # 2
  # 내가 작성
  # def __eq__(self, studentB): return super().getName() == studentB.getName() and super().getAge() == studentB.getAge() and self.__major == studentB.__major
  # 내가 상속받은 것과 내가 가지고 있는걸 구분해서 작성한거임 super() 와 studentB 사용
  # self.getName() == studentB.getName() and self.getAge() == studentB.getAge()
  # 부모 클래스에서 상속받은 메서드는 self.getName()처럼 self로도 호출 가능함
  # 단, 자식 클래스에서 메서드를 오버라이딩했을 경우 super()는 부모 메서드, self는 자식 메서드 호출

  # 강사님 답
  def __eq__(self, __o: object): # __o: object에서 object는 retrun 되는 타입 지정 / 속도 면에서도 좋음
    return super().__eq__(__o) & (self.__major == __o.__major) # 부모 클래스인 Person의 __eq__ 메서드 호출 & self와 __o 객체의 major 값이 같은지 비교

d = Student("Ko", 30, "computer science")
print(d)
d.setName("ku")
print("d = ", d)
e = Student("ku", 30, "computer science")
f = Student("Ko", 30, "computer science")
print("e = ", e)
print("f = ", f)
print("d == e",d == e)
print("d == f",d == f)
```
