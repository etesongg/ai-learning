```python
# 각 언어가 몇번 나오는지 딕셔너리로 만들기(내 풀이)
ch_cnt = {}
set_txt = set(txt_lst)
for c in set_txt:
  c_cnt = txt_lst.count(c)
  ch_cnt[c] = c_cnt # 딕셔너리 새로운 키-값 추가하기
print(ch_cnt)
```

```python
# 각 언어가 몇번 나오는지 딕셔너리로 만들기(강사님 풀이)
ch_cnt = {}
for c in txt:
  if c in ch_cnt.keys(): # 키값 중에 있다면/ 그냥 ch_cnt 딕셔너리 자체를 써도 가능은 하나 내부적으로 명시해 주는게 더 좋음, 보기도 좋음
    ch_cnt[c] += 1 # +1
  else:
    ch_cnt[c] = 1 # 없다면 1 집어넣기
print(ch_cnt)
```

```python
members = ['윤하원', '손형민', '오수성', '박지원', '조재홍',
           '이재원', '이상욱', '강호현', '정송희', '서민지',
           '이랑', '백지윤', '류효주', '이재현', '김종현',
           '이대현', '이효재', '두소진', '이득규', '김재헌',
           '김준영', '김영경']

# {'윤': ['윤하원'], '손': ['손형민'], '오': ['오수성'], '박': ['박지원'], '조': ['조재홍'], '이': ['이재원', '이상욱', '이랑',.. 출력하기
fn_names = {}
for name in members:
  # fn_names[name[0]] = [name]
  fn_names[name[0]] = fn_names.get(name[0], []) + [name] # '이': ['이재원', '이상욱', '이랑', ...
print(fn_names)
```

```python
# 위 결과에서 value가 가장 많은 값 출력하기

max_name = ''
max_len = 0
for key in fn_names:
  if max_len < len(fn_names[key]):
    max_len = len(fn_names[key])
    max_name = key
print(max_name, ":", fn_names[max_name])
```

```python
def top_ten_word(msgs):
  # 1. space를 기준으로 word 구분
  msg_split = msgs.split()

  # 2. 3글자 이하 제외
  valid_words = [] # 3글자 초과 단어 저장할 리스트
  for msg in msg_split:
    if len(msg) > 3:
      valid_words.append(msg)

  word_freq = {} # 단어별 빈도 수 저장할 딕셔너리
  for word in valid_words:
    word_freq[word] = word_freq.get(word, 0) + 1

  word_freq = sorted(word_freq.items(), key=lambda item:item[1], reverse=True)
  # .items()는 딕셔너리의 (key, value) 쌍을 튜플 형태로 반환
  # 정렬 기준을 지정, item은 각 (key, value) 쌍, item[1]은 그 중 value 값, value 기준으로 정렬
  # reverse=True 내림차순 정렬
  top_ten = word_freq[:10]

  return top_ten

top_ten_word(msg)
```

```python
# 오늘 배운걸로 짜기
def top_ten_word2(msgs):
  msg_split = msgs.split()

  valid_words = [] # 3글자 초과 단어 저장할 리스트
  for msg in msg_split:
    if len(msg) > 3:
      valid_words.append(msg)

  word_freq = {} # 단어별 빈도 수 저장할 딕셔너리
  for word in valid_words:
    word_freq[word] = word_freq.get(word, 0) + 1

  top_ten = [] # value 값 상위 10개 저장할 리스트
  i = 0
  word_freq_copy = word_freq.copy()
  while i< 10:
    max_word = ''
    max_count = 0

    for word in word_freq_copy:
      if max_count < word_freq_copy[word]: # key를 통해 value가 max_count보다 큰 값을 찾음
        max_count = word_freq_copy[word] # 해당 value 저장
        max_word = word # 해당 key 저장

    top_ten.append(max_word)
    del word_freq_copy[max_word] # 해당 값 삭제
    i += 1

  return top_ten

top_ten_word2(msg)
```

```python
# 1. space 기준으로 나누고
# 2. 3글자 이하 제외
# 3. 단어 빈도가 많은 상위 10개 출력

# 키 - 값 뒤집는 방법
# 오늘 배운걸로 짜기2
def top_ten_word3(msgs):
  # 1
  msgs_split = msgs.split()

  # 2
  vaild_words = []
  for msg in msgs_split:
    if len(msg) > 3:
      vaild_words.append(msg)
  print("vaild_words :",vaild_words)

  # 3-1 빈도 세기
  word_freq = {}
  for word in vaild_words:
    word_freq[word] = word_freq.get(word,0) + 1
  print("word_freq :", word_freq)

  # 3-2 빈도 기준 딕셔너리 만들기
  freq_to_words = {}
  for word in word_freq:
    freq_to_words[word_freq[word]] = freq_to_words.get(word_freq[word], []) + [word]
  print("freq_to_words :", freq_to_words)

  # 3-3 빈도가 많은 상위 10개 출력
  top_ten = []
  sort_freq = sorted(freq_to_words.keys(), reverse=True)
  print("sort_freq :", sort_freq)

  for freq in sort_freq:
    for word in freq_to_words[freq]:
      if len(top_ten) < 10:
        top_ten.append(word)
  print("top_ten :", top_ten)


top_ten_word3(msg)
```

```python
# 강사님 풀이
def top_ten_word(msg):
  top_ten = []
  word_cnt = {}
  for w in msg.split(): # split하면서 for문 돌려서 한번에 실행하기
    nw = w.lower() # 대소문자 따로 카운팅 하는걸 방지하기 위해 소문자로 만들기
    nw = nw.replace(".","")
    nw = nw.replace(",","")
    nw = nw.replace('"',"")
    nw = nw.replace("'s","")
    nw = nw.replace(":","")
    nw = nw.replace(";","")
    nw = nw.replace("?","")
    nw = nw.replace("!","")
    if not nw.isalpha(): # 알파벳이 아닌 다른 문자가 들어간 단어만 출력
      print(nw)
      continue # 해당 구역의 반복문 탈출 -> 여기에서는 바로 상위에 있는 for문으로 감

    if len(nw) <= 3:
      continue
    word_cnt[nw] = word_cnt.get(nw, 0) + 1
  print("word_cnt :",word_cnt)

  cnt_word = {}
  for w in word_cnt:
    cnt_word[word_cnt[w]] = cnt_word.get(word_cnt[w], []) + [w]
  print("cnt_word :",cnt_word)

  for cnt in sorted(cnt_word.keys(), reverse=True): # 키를 가져오는 것과 정렬 하는걸 한번에 실행
    if len(top_ten) <= 10:
      top_ten.extend(cnt_word[cnt]) # cnt_word[cnt]의 요소가 여러개이기 때문에 extend 사용
    else: # else break를 사용하는 이유는 for문을 끝까지 도는걸 막기 위해
      break
  print("top_ten :", top_ten)

  # .items() 튜플로 묶어서 리스트로 반환

  return top_ten


top_ten_word(msg)
```
