```python
# try-except 예외처리
def noname(seq1):
  try:
    seq1[0] = 'a'
  except: # try에서 에러발생하면 여기로 이동 / 문자열은 수정 안 됨
    # print("error : seq1 is string")
    pass
  print(seq1)
noname([1,2,3])
noname("abcd")
```

```python
# 이미지의 rgb color 채널의 평균 값?
r_mean = np.mean(ndimg[:, :, 0])
g_mean = np.mean(ndimg[:, :, 1])
b_mean = np.mean(ndimg[:, :, 2])

print(f'R color 평균: {r_mean}, G color 평균: {g_mean}, B color 평균: {b_mean}')
```

```python
# 이미지의 r,g,b color 채널별로 imshow 해보기
ndimg_r = ndimg.copy()
ndimg_r[:,:,1:] = 0 # 해당 컬러 외의 값 0 대입
plt.imshow(ndimg_r)
```

```python
ndimg_g = ndimg.copy()
ndimg_g[:,:,0::2] = 0 # 인덱스를 넣을때 (0,2)로 튜플로 작성가능 ndimg_g[:,:,(0, 2)]
plt.imshow(ndimg_g)
```

```python
ndimg_b = ndimg.copy()
ndimg_b[:,:,:2] = 0
plt.imshow(ndimg_b)
```

```python
# 이미지의 r,g, b 채널의 0.3, 0.35, 0.35 가중치로 흑백 이미지 만들어서 imshow 해보기
weights = [0.3, 0.35, 0.35]
gray_img = np.dot(ndimg, weights) # np.dot(ndimg_copy[:, :, :3], weights) 4인 경우가 있기 때문에 명시적으로 :3으로 적은거임 # ndimg_copy[...,:3]
plt.imshow(gray_img) # cmap='gray'
```

```python
# r 채널과 b채널을 바꿔서 imshow 해보기
change_rb = ndimg[:,:,[2,1,0]]

plt.imshow(change_rb)
```

```python
# 가로와 세로를 transpose 해서 imshow 해보기
trans_hv = ndimg.transpose((1,0,2))
plt.imshow(trans_hv)
```
