```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import glob

rootdir = '/content/drive/MyDrive/2025_SeSAC/python_programming/일반남여_일반통합06_F_1524374974_25_수도권_실내/'
flist = glob.glob(rootdir+'/*.json')

rec_time = []
rec_dt = []
for fname in flist:
  with open(fname) as f:
    json_data = json.load(f)

    rec_time.append(json_data['발화정보']['recrdTime'])
    rec_dt.append(json_data['발화정보']['recrdDt'])

print(rec_time)
print(rec_dt)
rec_time = np.array(rec_time, dtype=np.float16) # dtype 하지 않으면 str 형태임
rec_dt = np.array(rec_dt)
print(rec_time.shape)
print(rec_dt.shape)


plt.close()
fig = plt.figure(figsize=(6,6))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(rec_time)
ax2.hist(rec_time, bins=20, color='r', alpha=0.3)
# ax3.scatter(rec_time, rec_time + 3 * np.random.randn(10))
ax3.scatter(rec_time, rec_dt)
ax4.barh(rec_dt, rec_time) # 수평 막대 그래프

ax1.set_title('plot(rec_time)') # 제목 설정
ax1.set_xlabel('index') # x축 이름 설정
ax1.set_ylabel('sec') # y축 이름 설정
# fig.subplots_adjust(bottom=0.3) #
ticks = ax3.set_yticks([np.mean(rec_time)]) # y 눈금 설정
labels = ax3.set_yticklabels(['2021-01-05'], rotation=15, fontsize='small') # y 눈금의 라벨 설정, rotation 기울기

ticks = ax4.set_yticks([0])
labels = ax4.set_yticklabels(['2021-01-05'], rotation=15, fontsize='small')
```

fl2000.txt 를 dataframe 으로 읽은 뒤 다음을 실행하시오

Bush 와 Gore 의 총 득표수를 비교하는 그래프를 그리시오.

Bush 가 Gore 보다 많은 표를 받은 county top 10 개에 대해 득표수를 비교하는 그래프를 그리시오.

유권자가 많은 top 10 county 에 대해 Bush, Gore 의 득표수를 비교하는 그래프를 그리시오.

```python
import pandas as pd
rootdir = '/content/drive/MyDrive/2025_SeSAC/python_programming/'
fl2000 = pd.read_csv(rootdir+'fl2000.txt', delimiter='\s+', # 스페이스 구분자
                     skiprows=range(42), index_col='county')
fl2000.head()
```

```python
fl2000['total'] = fl2000.iloc[:,4:].sum(1) # 5번째 열부터 끝까지, sum(1) 행 기준
```

```python
fl2000.head() # 기본으로 5개만 나옴
```

```python
# Bush 와 Gore 의 총 득표수를 비교하는 그래프를 그리시오.
fl2000.sort_values('total', ascending=False)[['Bush','Gore']].sum(0).plot(kind='bar')
```

```python
# Bush 가 Gore 보다 많은 표를 받은 county top 10 개에 대해 득표수를 비교하는 그래프를 그리시오.
# fl2000.sort_values('total', ascending=False)[:10][['Bush','Gore']].plot(kind='bar') # total을 내림차순으로 정렬해서 상위 10개 슬라이싱 하고 'Bush','Gore' 만 나오게 한 후 그래프를 바 형태로 그려라
fl2000[fl2000['Bush']>fl2000['Gore']].sort_values('total', ascending=False)[:10][['Bush','Gore']].plot(kind='bar')
```

```python
# 유권자가 많은 top 10 county 에 대해 Bush, Gore 의 득표수를 비교하는 그래프를 그리시오.
fl2000.sort_values('total', ascending=False)[:10][['Bush','Gore']].plot(kind='bar')
```
