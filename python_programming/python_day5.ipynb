```python
# t는  숫자로 구성된 문자열 "3141592"
# p는 숫자로 구성된 문자열 "271"
# t를 p의 길이로 나누어 p보다 작은 수의 개수를 return
# t는 314, 141, 415, 592로 나누고 p인 217보다 작거나 같은 수의 개수는 2개

# 내 풀이
def soultion(t, p):
  answer = 0
  for i in range(len(t)-len(p)+1):
    if t[i:i+len(p)] <= p:
      answer += 1

  return answer

# soultion("3141592", "271")
soultion("500220839878", "7")
# soultion("10203", "15")
```

```python
# 강사님 풀이
def soultion(t, p):
  answer = 0
  for idx in range(len(t)-len(p)+1):
    if (int(t[idx:(idx+len(p))]) <= int(p)): # readability를 위해 () 묶어줌
      answer += 1
  return answer

print(soultion("3141592", "271") == 2)
print(soultion("500220839878", "7") == 8)
print(soultion("10203", "15") == 3)
```

```python
# 강사님 풀이
rootdir = "/content/drive/MyDrive/2025_SeSAC/"
fo = open(rootdir + 'score.txt', 'r')
txt = fo.readlines() # 현장에서는 파일의 용량이 크기 때문에 한번에 전체를 저장하는 것이 아닌 readline()을 사용하는게 일반적임

scores = {}
for line in txt:
  tmp = line.split()
  # scores[tmp[0]] = tmp[1:] # 범위이기 때문에 하나만 있어도 리스트로 나옴
  scores[tmp[0]] = list(map(int, tmp[1:])) # str이 아니라 int로 저장하고 싶다면? map 사용 하여 각각의 요소에 형식을 지정

fo.close()

print(scores)
```

```python
# kim와 가장 유사한 값 찾기 (유클리드 거리)
rootdir = "/content/drive/MyDrive/2025_SeSAC/"
fo = open(rootdir + 'score.txt', 'r')
txt = fo.readlines()

scores = {}
for line in txt:
  tmp = line.split()
  scores[tmp[0]] = list(map(int, tmp[1:]))

result = {}
kim = scores['kim']
for name in scores:
  if name != 'kim':
    diff = [(a - b) for a, b in zip(kim, scores[name])]
    dist = (diff[0]**2 + diff[1]**2)**0.5
    result[name] = dist
print(result)

min_dist = 100
min_name = ''
for name in result:
  if min_dist > result[name]:
    min_dist = result[name]
    min_name = name


fo.close()

print(min_name, min_dist)
```

```python
# kim와 가장 유사한 값 찾기 (유클리드 거리)
# 강사님 풀이
p1 = scores['kim']
min_dist = 20001 # 차이가 많이 나봤자 100이기 때문에 100**2 + 100**2의 값 + 1을 넣은거임
c_n = ''
for n in scores.keys():
  if n != 'kim':
    p2 = scores[n]
    dist = 0
    for idx in range(len(p1)):
      dist += (p1[idx] - p2[idx])**2 # 차의 제곱의 합
    if dist < min_dist:
      min_dist = dist
      c_n = n

print(c_n, min_dist**0.5)
```
