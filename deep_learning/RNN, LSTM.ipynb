```python
txt = 'hihello'
set(txt)
```

```python
ch2id = {ch:id for id, ch in enumerate(set(txt))}
print(ch2id)
txt_id = [ch2id[ch] for ch in txt]
print(txt_id)
```

```python
from tensorflow.keras.utils import to_categorical
data = to_categorical(txt_id)
print(data)
```

```python
import numpy as np
x_data = data[:-1][np.newaxis,:,:] # hihell
y_data = data[1:][np.newaxis,:,:] # ihello
print(x_data.shape, y_data.shape)
```

```python
# 모델 만들기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, SimpleRNN, LSTM

hihello_model = Sequential()
hihello_model.add(Input(shape=(6, 5))) # (배치,) 갯수, 크기
hihello_model.add(LSTM(units=16, return_sequences=True))
hihello_model.add(Dense(5, activation='softmax')) # Dense를 통해 shape 맞춤
hihello_model.summary()

hihello_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
hihello_model.fit(x_data, y_data, epochs=100)
```

```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape, y_test.shape)
```

```python
np.max(x_train)
```

```python
x_train_norm = x_train / 255
x_test_norm = x_test / 255

y_train_oh = to_categorical(y_train)
y_test_oh = to_categorical(y_test)
print(np.max(x_train_norm), y_test_oh.shape)
```

```python
x_train_norm.shape, y_train_oh.shape
```

```python
# 모델 만들기 (LSTM의 결과가 잘 나오면 데이터의 성격이 시퀀스라는 걸 알 수 있음.)
mnist_lstm = Sequential()
mnist_lstm.add(Input(shape=(28, 28)))
mnist_lstm.add(LSTM(units=32)) # 마지막 시점의 출력만 반환
mnist_lstm.add(Dense(10, activation='softmax'))
mnist_lstm.summary()
mnist_lstm.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
mnist_lstm.fit(x_train_norm, y_train_oh,
               validation_data=(x_test_norm, y_test_oh),
               epochs=2)
```

```python
docs = ['나는 당신을 사랑합니다', '나는 당신을 미워합니다', 'i love you', 'i hate you']

wdlst = {}
for sent in docs:
  wdlst = set(wdlst).union(set(sent.split(' ')))
print(wdlst)
wd2id = {wd:id for id, wd in enumerate(wdlst)}
id2wd = {id:wd for id, wd in enumerate(wdlst)}
print(wd2id)
print(id2wd)

x_data=[]
for sent in docs:
  x_data.append([wd2id[wd] for wd in sent.split(' ')])
print(x_data)
x_data_oh = to_categorical(x_data)
print(x_data_oh.shape)
print(x_data_oh)
```

```python
translate_model = Sequential()
translate_model.add(Input(shape=(3, 8)))
translate_model.add(LSTM(32, return_sequences=True))
translate_model.add(Dense(8, activation='softmax'))
translate_model.summary()

translate_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# 한글 -> 영어
translate_model.fit(x_data_oh[:2], x_data_oh[2:], epochs=100)
pred = translate_model.predict(x_data_oh[:2])
# 영어 -> 한글
# translate_model.fit(x_data_oh[2:], x_data_oh[:2], epochs=100)
# pred = translate_model.predict(x_data_oh[2:])
print(np.argmax(pred, axis=2))
```

- 현재 모델은 정확도가 50%임 그 이유는 영어와 한글의 어순차이 때문임
- 그렇기 때문에 한글 -> 영어는 정확도가 낮지만
- 영어 -> 한글로 하면 정확도가 높음

```python
from tensorflow.keras.layers import Bidirectional

translate_model = Sequential()
translate_model.add(Input(shape=(3, 8)))
translate_model.add(Bidirectional(LSTM(32, return_sequences=True)))
translate_model.add(Dense(8, activation='softmax'))
translate_model.summary()

translate_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
translate_model.fit(x_data_oh[:2], x_data_oh[2:], epochs=100)
pred = translate_model.predict(x_data_oh[:2])
print(np.argmax(pred, axis=2))
```
