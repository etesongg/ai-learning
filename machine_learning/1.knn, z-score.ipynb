# KNN 알고리즘
iris 데이터 사용

```python
from sklearn.datasets import load_iris # iris 데이터 불러오기
iris = load_iris()
print(iris.DESCR)
```

```python
data = iris.data
target = iris.target # labels
print(data)
print(target)
```

```python
print(data.shape, target.shape)
# :Attribute Information:
#     - sepal length in cm
#     - sepal width in cm
#     - petal length in cm
#     - petal width in cm
#     - class:
#             - Iris-Setosa
#             - Iris-Versicolour
#             - Iris-Virginica
```

```python
import numpy as np
np.unique(target, return_counts=True) # return_counts=True 각각의 유니크 값이 target에서 몇 번 나왔는지 확인
```

```python
from sklearn.model_selection import train_test_split # 데이터를 학습용과 테스트용 데이터로 나누는 함수
x_train, x_test, y_train, y_test = train_test_split(data, # 입력 데이터(꽃의 특징)
                                                    target, # 정답 레이블(꽃의 종류)
                                                    test_size=0.3, # 테스트 데이터셋의 크기를 결정, 0.25가 기본
                                                    stratify=target, # 모집단의 클래스 비율 유지 설정, target의 클래스 비율을 학습/테스트 데이터에도 동일하게 유지
                                                    random_state=0) # 랜덤 시드 설정 (결과 재현 가능성 확보용)
print(f"x_train : {x_train.shape}, y_test : {y_test.shape}")
```

```python
np.unique(y_test, return_counts=True) # 클래스 비율이 고르게 분포되어 있지 않은 걸 확인(bias), 모델 평가가 왜곡될 가능성 있음
```

```python
np.unique(y_test, return_counts=True) # stratify=target를 통해 모집단의 클래스 비율 유지 설정했기 때문에 array([15, 15, 15] 나옴
```

KNN에는 Z-score 표준화가 더 적합

Min-Max Normalization는 range를 벗어날 경우가 생길 수 있으며, 이상치 혹은 훈련 범위를 벗어난 새로운 데이터에 매우 민감

```python
print(np.min(x_train, axis=0)) # 현재 shape(150, 4), 나는 4만 남는걸 원하기 때문에 = 열에 대한 값을 원하기 때문에 axis=0을 해줌
print(np.max(x_train, axis=0))
print(np.mean(x_train, axis=0))
print(np.std(x_train, axis=0))
```

z-score 스케일링 (표준화, Standardization) 데이터의 평균을 0, 표준편차를 1로 변환

직접 구현한 z-score 스케일링
- StandardScaler().fit(x_train).transform(x_train) 와 동일한 결과

```python
(x_train - np.mean(x_train, axis=0)) / np.std(x_train, axis=0)
# x_train (105,4), np.mean(x_train, axis=0) (4,) -> (105, 4)
# (값 - 평균) / 표준편차
```

StandardScaler를 활용한 z-score 스케일링

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train) # 데이터의 평균과 표준편차를 계산하여 저장

# fit에서 계산된 평균과 표준편차를 사용하여 x_train 데이터를 z-score 방식으로 표준화함
x_train_zst = scaler.transform(x_train)
# test는 fit을 사용하지 않는 이유 : test는 미래에 나올만한 데이터이기때문에 평균 표준편차를 모름, 데이터 누설을 방지
x_test_zst = scaler.transform(x_test)

print(x_train[:10])
print(x_train_zst[:5])
```

KNN 알고리즘

```python
from sklearn.neighbors import KNeighborsClassifier
# 모델 생성
iris_knn = KNeighborsClassifier(n_neighbors=10) # 사용할 이웃의 수
# 모델 학습
iris_knn.fit(x_train_zst, y_train) # (z-score로 표준화된 훈련 데이터, 해당 데이터의 라벨)

# 예측
pred = iris_knn.predict(x_test_zst)
```

```python
print(y_test)
print(pred)
```

```python
print(np.mean(y_test == pred))
```
