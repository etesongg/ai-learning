```python
import numpy as np

data = np.array([[0,0],[0,1],[1,0],[1,1]])
target = np.array([0,0,0,1])
weights = np.array([-0.3, 0.4, 0.1])
learning_rate = 0.01

for epoch in range(10):
  for x, y in zip(data, target):
    # print(x, y)
    y_hat = np.sum(x*weights[1:]) + weights[0]
    # print(y_hat)
    if y_hat >= 0:
      result = 1
    else:
      result = 0
    # print(result)
    error = y - result
    # print(error)
    if error != 0:
      # weights[0]  += learning_rate * error
      # weights[1:] += learning_rate * error * x

      weights[0] = weights[0] + learning_rate * error * 1
      weights[1:] = weights[1:] + learning_rate * error * x

  print(epoch + 1, "epoch, weights :", weights)
```

```python
from sklearn.neural_network import MLPClassifier # Multi-layer Perceptron classifier
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()

x_train, x_test, y_train, y_test = train_test_split(iris.data,
                                                    iris.target,
                                                    stratify=iris.target,
                                                    random_state=7,
                                                    test_size=0.3)

scaler = StandardScaler()
scaler.fit(x_train)
x_train_zs = scaler.transform(x_train)
x_test_zs = scaler.transform(x_test)

iris_nn = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=5000) # 히든 레이어 한장 =(10,) 현재 두장
iris_nn.fit(x_train_zs, y_train)
pred = iris_nn.predict(x_test_zs)
print(accuracy_score(y_test, pred))
```

**내 풀이**

- MLPClassifier의 hidden_layer_sizes 값에 따라서 어떤 size가 제일 성능이 좋을까?

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

x_train = train.drop(columns='Activity')
y_train = train['Activity']

x_test = test.drop(columns='Activity')
y_test = test['Activity']

# 문자열 레이블 숫자로 변환
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', MLPClassifier(max_iter=1000, early_stopping=True, random_state=7)) # early_stopping=True 수렴하면 학습 조기 종료
])

param_grid = {
    'model__hidden_layer_sizes': [
        (10,),
        (50,),
        (100,),
        (10, 10)
    ]
}

grid = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(x_train, y_train_encoded)

print("Best structure:", grid.best_params_)
print("Best CV accuracy:", grid.best_score_)

y_pred = grid.predict(x_test)
print("Test accuracy:", accuracy_score(y_test_encoded, y_pred))
```

- MLPClassifier의 hidden_layer_sizes (50,) 기준 제일 성능이 좋았음
- 그렇다면 다른 모델들과 비교했을때 어떤 모델이 제일 성능이 좋을까?

MLPClassifier의 hidden_layer_sizes (50,) 과

LogisticRegression, RandomForestClassifier, AdaBoostClassifier 모델의 성능비교

해당 모델 선택 이유:

- LogisticRegression: 기본적인 선형 분류 모델로, MLP(다층 퍼셉트론)와 비교하여 비선형 모델이 필요한 문제인지 판단하기 위해 선택

- RandomForestClassifier: 트리 기반의 비선형 모델로, 복잡한 데이터 패턴을 포착하는 MLP와의 성능 차이를 비교하고자 선택

- AdaBoostClassifier: 다른 방식의 앙상블 학습 모델로, MLP와 비교하여 다양한 학습 접근 방식의 효과를 검토하기 위해 선택



```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier

x_train = train.drop(columns='Activity')
y_train = train['Activity']

x_test = test.drop(columns='Activity')
y_test = test['Activity']

# 레이블 인코딩
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', MLPClassifier())
])

param_grid = [
    {
        'model': [MLPClassifier(max_iter=1000, early_stopping=True, random_state=7)],
        'model__hidden_layer_sizes': [(50,)],
    },
    {
        'model': [LogisticRegression(max_iter=1000)],
    },
    {
        'model': [RandomForestClassifier()],
        'model__n_estimators': [50, 100],
        'model__max_depth': [1, 5, 10]
    },
    {
        'model': [AdaBoostClassifier()],
        'model__n_estimators': [50, 100],
        'model__learning_rate': [0.5, 1.0]
    }
]

grid = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(x_train, y_train_encoded)

print("Best structure:", grid.best_params_)
print("Best CV accuracy:", grid.best_score_)

y_pred = grid.predict(x_test)
print("Test accuracy:", accuracy_score(y_test_encoded, y_pred))
```
